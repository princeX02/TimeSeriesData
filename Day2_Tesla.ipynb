```python
import yfinance as yf
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
```

```python
tsla_data=yf.download('TSLA',start='2020-01-01',end='2025-01-01')
```

```python
tsla_data
```

```python
#Simple Moving Average
```

```python
tsla_data['Open'].plot(xlim=['2021-01-01','2025-01-01'],ylim=[100,500],figsize=(15,8))
```

```python
tsla_data['10 Days Rolling']=tsla_data['Open','TSLA'].rolling(window=10).mean()
```

```python
tsla_data[['Open','10 Days Rolling']].plot(xlim=['2024-01-01','2025-01-01'],figsize=(13,8))
```

```python
tsla_data['20 Days Rolling']=tsla_data['Open','TSLA'].rolling(window=20).mean()
```

```python
tsla_data[['20 Days Rolling','Open']].plot(xlim=['2024-01-01','2025-01-01'],figsize=(15,8))
```

```python
#Expanding
#CMA
```

```python
tsla_data['Open'].expanding().mean().plot(figsize=(15,8))
```

```python
#EWMA Exponential Weighted Moving Average
```

```python
#EMA Exponential Moving Average 
# Removing lag
```

```python
tsla_data['EMA_10']=tsla_data['Open'].ewm(alpha=0.1,adjust=False).mean()
```

```python
tsla_data
```

```python
tsla_data[['Open','EMA_10']].plot(figsize=(15,8))
```

```python
tsla_data['EWMA_10']=tsla_data['Open','TSLA'].ewm(span=5).mean()
```

```python
tsla_data[['Open','EWMA_10']].plot(figsize=(15,8))
```
